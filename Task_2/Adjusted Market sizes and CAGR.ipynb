```python
import random
m_random= []
f_random=[]
cagr_random=[]

# Set a length of the list to 10
for i in range(0, 99):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    x = round(random.uniform(0.10, 0.50), 2)
    cagr_random.append(x)

for i in range(0, 99):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    x = round(random.uniform(0.01, 0.9), 1)
    m_random.append(x)

for i in range(0, 99):
    # any random float between 50.50 to 500.50
    # don't use round() if you need number as it is
    x = round(random.uniform(0.01, 0.9), 2)
    f_random.append(x)
```

```python
import pandas as pd
df=pd.read_csv('Beauty_PC.csv',encoding='Latin-1')
m_size=df['Market size'].to_list()
f_size=df['Forecasted Market Size/ Revenue'].to_list()
cagr=df['CAGR (%) '].to_list()
```

```python
all_billion=[]
for i in range(0,len(m_size)):
  text=m_size[i] 
  if (text!='0') and (text.find('million')!=-1):
    d=float(text[4:-9])
    billion=round(d/1000,4) 
    all_billion.append(billion)
  elif (text!='0') and (text.find('million')==-1):
    d1=float(text[4:-9])
    all_billion.append(d1)
  elif text=='0':
    all_billion.append(float(0))
```

```python
f_all_billion=[]
for i in range(0,len(f_size)):
  text=f_size[i] 
  if (text!='0') and (text.find('million')!=-1):
    d=float(text[4:-9])
    billion=round(d/1000,4) 
    f_all_billion.append(billion)
  elif (text!='0') and (text.find('million')==-1):
    d1=float(text[4:-9])
    f_all_billion.append(d1)
  elif text=='0':
    f_all_billion.append(float(0))
```

```python
#adjust_m_size=['USD '+str(round(float(i+j),3))+' billion' for i in m_random for j in all_billion]
adjust_m_size=[]
for i in m_random: 
  for j in all_billion:
    if j == 0.0:
      adjust_m_size.append('NA')
    else:
      final='USD '+str(round(float(i+j),3))+' billion'
      adjust_m_size.append(final)
```

```python
adjust_f_size=[]
for i in f_random: 
  for j in f_all_billion:
    if j == 0.0:
      adjust_f_size.append('NA')
    else:
      final='USD '+str(round(float(i+j),3))+' billion'
      adjust_f_size.append(final)
```

```python
adjust_f_size
```

```python
cagr_val=[]
for i in range(0,len(cagr)):
  n=cagr[i].replace("%","")
  fl=float(n)
  cagr_val.append(fl)
```

```python
adjust_cagr=[]
for i in cagr_random: 
  for j in cagr_val:
    if j == 0.0:
      adjust_cagr.append('NA')
    else:
      final=str(round(float(i+j),3))+'%'
      adjust_cagr.append(final)
```

```python
adjust_cagr
```

```python
data={'Market Size':adjust_m_size, 'Forecasted Market Size':adjust_f_size,'CAGR (%)':adjust_cagr}
df1=pd.DataFrame(data=data)
df1.to_csv('Adjusted-BPC.csv',index=False)
```
